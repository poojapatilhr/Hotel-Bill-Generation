```python
def getTax(totol_Bill):
    return totol_Bill * (10.1 / 100)

def print_item(num, n, p):
    num_of_spaces = num - len(n)
    result = num_of_spaces * " " + n + "  " + "${:,.2f}".format(p) + "\n"
    return result

import time
import datetime

def printReceipt(order_List, A):
    
    subTotal = 0
    # Iterate over each order and calculate subtotal
    for order in order_List:
        choice = order[0]
        quantity = order[1]
        subTotal += A[choice][1] * quantity
    
    # Get tax
    tax = getTax(subTotal)
    
    # Calcualte total bill
    total_Bill = subTotal + tax
    
    # Generate receipt
    
    print()
    print()
    print ("Date: ", datetime.datetime.now())
    print("----------------------------------")
    
    # Print ordered menu
    for order in order_List:
        choice = order[0]
        quantity = order[1]
        order_Item_Name = A[choice][0]
        
        num_of_spaces = 20 - len(order_Item_Name) - 1 
        num_of_spaces2 = 5 - len(str(quantity))
        print(num_of_spaces * " " + order_Item_Name + ":  " + str(quantity) + num_of_spaces2 * " " + "$" + str(A[choice][1]))
    
    print()
    print()
    print(print_item(20,'SUBTOTAL:', subTotal))
    print(print_item(20,'TAX:', tax))

    print("----------------------------------")
    print(print_item(20,'TOTAL DUE:', total_Bill))      

    print()
    print()
    print("THANK YOU FOR DINING WITH US!")
    print("PLEASE COME AGAIN")
    
```
